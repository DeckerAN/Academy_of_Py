### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())

total_players_df = pd.DataFrame({"Total Players": [total_players]})

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of Unique Items
num_unique_items = len(purchase_data["Item ID"].unique())

# Average Purchase Price
avg_price = purchase_data["Price"].mean()

# Total Number of Purchases
total_purchases = purchase_data["Purchase ID"].count()

# Total Revenue
total_rev = purchase_data["Price"].sum()

# Give the displayed data cleaner formatting
cln_avg_price = f"${avg_price:.2f}"
cln_total_rev = f"{total_rev:,.2f}"

# Display the summary data frame
purchase_summary_df = pd.DataFrame({"Number of Unique Items": [num_unique_items],
                                    "Average Purchase Price": [cln_avg_price],
                                    "Total Number of Purchases": [total_purchases], 
                                    "Total Revenue": [cln_total_rev]
                                   })

purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed

```python
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

unique_sn_max_df = purchase_data.groupby("SN").max()

gender_group_df = unique_sn_max_df[["Gender", "Purchase ID"]].groupby("Gender").count()

gender_group_df = gender_group_df.rename(columns={"Purchase ID": "Total Count"})

gender_group_df["% of Players"] = (gender_group_df["Total Count"] / total_players) * 100

gender_group_df["% of Players"] = gender_group_df["% of Players"].map("{:.2f}".format)

del gender_group_df.index.name

gender_group_df.sort_values("Total Count", ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase = purchase_data.groupby("Gender")

# Purchase Count
gen_purchase_count = (gender_purchase.count())["Purchase ID"]

# Display the summary data frame
gender_purchase_df = pd.DataFrame(gen_purchase_count)
gender_purchase_df = gender_purchase_df.rename(columns={"Purchase ID": "Purchase Count"})

# Average Purchase Price
gender_purchase_df["Avg Purchase Price"] = (gender_purchase.mean())["Price"]

# Total Purchase Value
gender_purchase_df["Total Purchase Value"] = (gender_purchase.sum())["Price"]

# Average Purchase Total per Person by Gender
gender_purchase_df["Avg Total Purchase per Person"] = (gender_purchase.sum())["Price"] / gender_group_df["Total Count"]

# Give the displayed data cleaner formatting
gender_purchase_df["Avg Purchase Price"] = gender_purchase_df["Avg Purchase Price"].map("${:.2f}".format)
gender_purchase_df["Total Purchase Value"] = gender_purchase_df["Total Purchase Value"].map("{:,.2f}".format)
gender_purchase_df["Avg Total Purchase per Person"] = gender_purchase_df["Avg Total Purchase per Person"].map("${:.2f}".format)

del gender_purchase_df.index.name
gender_purchase_df = gender_purchase_df.sort_values("Avg Total Purchase per Person", ascending=False)

gender_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 116]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create a summary data frame to hold the results
# Categorize the existing players using the age bins. Hint: use pd.cut()
unique_sn_mean_df = purchase_data.groupby("SN").mean()
sn_binned_ages = pd.cut(unique_sn_mean_df["Age"], age_bins, labels=age_labels)

# Calculate the Total Count...
sn_binned_ages = sn_binned_ages.reset_index()
binned_ages_df = sn_binned_ages.groupby("Age").count()
binned_ages_df = binned_ages_df.rename(columns={"SN": "Total Count"})

# ...and percentages by age group
# Optional: round the percentage column to two decimal points
binned_ages_df["% of Players"] = round((binned_ages_df["Total Count"] / total_players) * 100, 2)

# Display Age Demographics Table
binned_ages_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data data frame by age
purchase_data["Age Bins"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)

purchase_ages_group = purchase_data.groupby("Age Bins")

# Purchase Count
purchase_ages_df = pd.DataFrame((purchase_ages_group.count())["Purchase ID"])
purchase_ages_df = purchase_ages_df.rename(columns={"Purchase ID": "Purchase Count"})

# Average Purchase Price
purchase_ages_df["Avg Purchase Price"] = (purchase_ages_group.mean())["Price"]

# Total Purchase Value
purchase_ages_df["Total Purchase Value"] = (purchase_ages_group.sum())["Price"]

# Average Purchase Total per Person by Age Group
purchase_ages_df["Avg Total Purchase per Person"] = purchase_ages_df["Total Purchase Value"] / binned_ages_df["Total Count"]

# Optional: give the displayed data cleaner formatting
purchase_ages_df["Avg Purchase Price"] = purchase_ages_df["Avg Purchase Price"].map("${:.2f}".format)
purchase_ages_df["Total Purchase Value"] = purchase_ages_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_ages_df["Avg Total Purchase per Person"] = purchase_ages_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchase_ages_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
## Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

sn_group = purchase_data.groupby("SN")

# SN
unique_sn_count_df = sn_group.count()

# Purchase Count
top_spenders = pd.DataFrame(unique_sn_count_df["Purchase ID"])
top_spenders = top_spenders.rename(columns={"Purchase ID": "Purchase Count"})

# Average Purchase Price
top_spenders["Average Purchase Price"] =  (sn_group.sum())["Price"] / unique_sn_count_df["Purchase ID"]


# Total Purchase Value
top_spenders["Total Purchase Value"] = (sn_group.sum())["Price"]

# Sort the total purchase value column in descending order
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

# Optional: give the displayed data cleaner formatting
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)

top_spenders.head(5)
```
